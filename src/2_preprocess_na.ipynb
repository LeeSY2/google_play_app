```python
import pandas as pd

from lib import get_processing_folder_path
```

```python
# df = pd.read_csv(get_file_path('Google-Playstore.csv'))
df = pd.read_csv(get_processing_folder_path('processed_drop.csv'))
```

```python
df.isna().sum()
```

```python
# drop na 
# App Name은 해당 애플리케이션 찾아서 직접 삽입
# 다음 컬럼의 NA 값은 drop - Rating, Rating Count, Released, Currency, Minimum Android
# Rating and Rating Count is 0 은 확인 후 남기기로 결정
```

```python
# App Name
df[df['App Name'].isna()]
```

```python
target = df.loc[df['App Id'].str.match('com.Tomkii.NULl')].index.values[0]
df.loc[target,'App Name'] = 'NULL APK'

target2 = df.loc[df['App Id'].str.match('se.promedia.NA')].index.values[0]
df.loc[target2,'App Name'] = 'NA APK'

df[df['App Name'].isna()].shape[0]
```

```python
# Rating & Rating Count

# 두 na 값들이 서로 같은 record인지 확인
# 각각 22883개

total = 22883
rating = df.loc[df['Rating'].isna()].index
rating_count = df.loc[df['Rating Count'].isna()].index

(rating == rating_count).sum() == total
```

```python
drop_list = ['Rating','Rating Count']
drop_r_rc = df.dropna(subset=drop_list)

drop_r_rc.to_csv(get_processing_folder_path('midterm_processed.csv'),index=False)
drop_r_rc.isna().sum()
```

```python
# drop Minimum Android column
drop_ma = drop_r_rc.drop(labels=['Minimum Android'],axis=1)
drop_ma.isna().sum()
```

```python
# Currency
## XXX 는 체코슬로바키아 화폐단위
drop_ma['Currency'].unique()
c_na = drop_ma[drop_ma['Currency'].isna()]
c_na
```

```python
drop_c = drop_ma.dropna(subset=['Currency'])
drop_c.loc[drop_c.index.isin(c_na.index)]
```

```python
drop_c.isna().sum()
```

```python
drop_r = drop_c.dropna(subset=['Released'])
drop_r.isna().sum()
```

```python
drop_r
```

```python
tf_dt = drop_r.copy(deep=True)

tf_dt.loc[:,'Released'] = pd.to_datetime(tf_dt['Released'])
```

```python
tf_dt.loc[:,'Last Updated'] = pd.to_datetime(tf_dt['Last Updated'])
```

```python
tf_dt.loc[:,'Minimum Installs'] = tf_dt['Minimum Installs'].astype(int)
```

```python
tf_dt.loc[:,'Rating Count'] = tf_dt['Rating Count'].astype(int)
```

```python
tf_dt.info()
```

```python
tf_dt.to_csv(get_processing_folder_path('processed_na.csv'),index=False)
```
